```python
imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d1[['Result']])
d1['Result'] = imputer.transform(d1[['Result']])    

imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d1[['When Scored']])
d1['When Scored'] = imputer.transform(d1[['When Scored']])    

```

```python
imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d2[['Result']])
d2['Result'] = imputer.transform(d2[['Result']])    

imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d2[['When Scored']])
d2['When Scored'] = imputer.transform(d2[['When Scored']])    
```

```python
imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d3[['Result']])
d3['Result'] = imputer.transform(d3[['Result']])    

imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d3[['When Scored']])
d3['When Scored'] = imputer.transform(d3[['When Scored']])    
```

```python
imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d4[['Result']])
d4['Result'] = imputer.transform(d4[['Result']])    

imputer = SimpleImputer(strategy='most_frequent')
imputer = imputer.fit(d4[['When Scored']])
d4['When Scored'] = imputer.transform(d4[['When Scored']])    
```
